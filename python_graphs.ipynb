```
import matplotlib.pyplot as plt
import numpy as np
import time
```

```
x=[4,7,8,9,11,5]
y=[13,23,5,7,9,4]
player=['Virat','Dhoni','Vijay','Dhawan']
runs=[120,44,77,33]
exp=[0.1,0,0,0.2]
```

```
#@title Bar Graph { vertical-output: true }


plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(color='green')

plt.bar(x,y,label='pollution')

plt.bar(player,runs,label='cricketscore') 
plt.legend()
plt.show()

```

```
#@title Dotted Graph { vertical-output: true }

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(color='green')
plt.scatter(x,y) #only points
plt.legend()
plt.show()
```

```
#@title Pie Chart { vertical-output: true }
plt.pie(runs,labels=player,explode=exp,shadow=True)

plt.show()
```

```
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(color='green')
plt.plot(x,y) 
plt.legend()
plt.show()
```
